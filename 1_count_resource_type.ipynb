```python
import codecs
from glob import glob
import json
from collections import Counter

%store -r resource_types
```

```python
files = glob("json_files/*", recursive = True)
```

```python
resource_key_list = {}
for resource_type in resource_types:
    print(resource_type)
    not_match = False
    key_lists = []
    for one_file in files:
        with codecs.open(one_file, encoding='utf-8', errors='ignore') as f:
            d = json.load(f)
            for i, obj in enumerate(d['entry']):
                if obj['resource']['resourceType'] == resource_type:
                    resource_obj = obj['resource']
                    current_keys = ', '.join(sorted(list(resource_obj.keys())))
                    key_lists.append(current_keys)
    resource_key_list[resource_type] = Counter(key_lists)
```
